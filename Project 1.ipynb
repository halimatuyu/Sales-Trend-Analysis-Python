# Sales Trend

```python
#MIN DATE

data['order_date'].min()
```

```python
#MAX DATE

data['order_date'].max()
```

```python
#Getting out the Year and the Month from the dataset

data['month_year'] = data['order_date'].apply(lambda x: x.strftime('%Y-%M'))
```

```python
data['month_year']
```

```python
#Grouping by month year

data_trend = data.groupby('month_year').sum()['sales'].reset_index()
```

# Plotting

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data_trend['month_year'], data_trend['sales'], color='red')
plt.show()
```

# Top 10 product by sales
